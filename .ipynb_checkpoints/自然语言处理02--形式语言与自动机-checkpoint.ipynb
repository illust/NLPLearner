### 2. 形式语言  
一般描述一种语言可以有三种途径：
1. 穷举法：把语言中的所有句子都枚举出来。这种方法只适合句子数目有限的语言。
2. 文法（产生式系统），英文为grammar，描述：语言中的每个句子用严格定义的规则来构造，利用**规则**生成语言中合法的句子。  
3. 自动机法：通过对输入的句子进行合法性检验，区别哪些是语言中的句子，哪些不是语言中的句子。  

**教材摘要：**
>文法用来精确地描述语言和其结构，自动机则是用来机械地刻画对输入字符串的识别过程。  
用文法来定义语言的优点：由文法给予语言中的句子以结构，各成分之间的结构关系清楚，明了。但是，如果要直接用这些规则来确定一个字符串是否属于这套规则所定义的语言似乎并不十分明确。而由自动机来识别一个字符串是否属于该语言则相对简单，但自动机很难描述语言的结构。所以自然语言处理中的识别和分析算法，大多兼取两者之长。  

#### 2.1 形式语言的类型  
乔姆斯基理论中，文法被分为四种类型：3型文法，2型文法，1型文法和0型文法，分别称为正则文法（也称有限状态文法），上下文无关文法，上下文相关文法和无约束文法。

### 3. 自动机理论  
自动机分类：有限自动机，下推自动机，线性界限自动机和图灵机。
#### 3.1 有限自动机  
- 确定性有限自动机(DFA)
- 不确定性有限自动机(NFA)  

简单来说，有限自动机主要用来识别正则语言和上下文无关语言。从起始状态开始，通过状态转换函数（通过现在的状态和下一个字符来决定去哪个状态）转移到下一个状态，直到读完所有的字符，如果最后的状态是accepting，则表示这个字符串是属于我们所规定的语言，否则不属于。  
DFA和NFA的区别主要是状态转移函数不同。对于DFA，一个特定的符号输入，有且只能得到一个状态，而NFA就有可能得到一个状态集。
